```python
import pandas as pd
import seaborn as sns
import scipy.stats as stats

sns.set_style("whitegrid")
```

```python
data_full = pd.read_csv("data_full.csv", index_col=0)
```

## Chart generation

```python
g = sns.catplot(x='experiment', y='hmean', hue='approach', data=data_full, kind='bar', ci='sd')
g.set_xticklabels(['Experiment 1', 'Experiment 2', 'Experiment 3'])
g.set_ylabels('H-Mean')
g.set_xlabels('')
g.legend.set_title('Approach')
g.legend.set_bbox_to_anchor((1, 0.5))
g.legend.set_frame_on(True)
g.tight_layout()
g.savefig('metric_comparison.pdf')
```

## ANOVA Test

```python
anovas = pd.DataFrame(columns=['experiment', 'stat_par', 'eq_odds', 'zero_one_loss', 'disp_imp', 'acc', 'hmean'])
```

## Experiment 1

```python
exp1 = data_full[data_full['experiment'] == 'exp1']
```

```python
exp1
```

```python
anovas.loc[0, 'experiment'] = 'Experiment 1'
```

```python
for c in exp1.drop(columns=['experiment', 'approach']).columns:
  f_value, p_value = stats.f_oneway(exp1.query('approach == "original"')[c].abs(), exp1.query('approach == "MANILA"')[c].abs())
  #print(p_value)
  anovas.loc[0, c] = p_value
```

```python
anovas
```
